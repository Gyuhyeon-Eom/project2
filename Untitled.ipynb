# 취소 거래 데이터 분석!

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 구매 시간을 시각화하기 위해, 'InvoiceDate'에서 시간을 'Hour'로 변환합니다.
cancel_data_subset['Hour'] = pd.to_datetime(cancel_data_subset['InvoiceDate']).dt.hour

# 시각화 준비
plt.figure(figsize=(10, 6))

# 시간대별로 취소된 거래의 수를 시각화 (히스토그램)
sns.histplot(cancel_data_subset['Hour'], bins=24, kde=True)
plt.title('Cancelled Transactions by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Cancelled Transactions')

# 그래프 출력
plt.show()
```

```python
# 구매 'year와 month'를 추출하기 위해 'InvoiceDate' 컬럼에서 'yyyy-mm' 정보를 가져옵니다.
cancel_data_subset['yyyy-mm'] = pd.to_datetime(cancel_data_subset['InvoiceDate']).dt.to_period('M')

# 필요한 패키지 불러오기
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 시각화 준비
plt.figure(figsize=(10, 6))

# 'yyyy-mm' 기준으로 취소된 거래 수 시각화 (연도-월별 취소 건수)
cancel_counts = cancel_data_subset.groupby('yyyy-mm').size()

# 시각화
cancel_counts.plot(kind='bar', color='skyblue')
plt.title('Cancelled Transactions by Year-Month')
plt.xlabel('Year-Month')
plt.ylabel('Number of Cancelled Transactions')

# 그래프 출력
plt.tight_layout()
plt.show()
```

### cancel 데이터의 가격 합

```python
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv('online_retail_II.csv.zip')

cancel_data = df[df['Invoice'].astype(str).str.contains('C', case=False)]

# 'Price'의 합
cancel_price_sum = cancel_data['Price'].sum()

cancel_price_sum
```

```python
# 데이터에서 'Invoice' 컬럼에 'C'가 없는 정상 거래
valid_sales_data = df[~df['Invoice'].astype(str).str.contains('C', case=False)]

# 정상 거래에서 'Price'의 합
valid_sales_price_sum = valid_sales_data['Price'].sum()

valid_sales_price_sum
```

```python
# 취소된 거래의 'Price' 합과 정상 거래의 'Price' 합을 구합니다.
cancel_price_sum = cancel_data['Price'].sum()
valid_sales_price_sum = valid_sales_data['Price'].sum()

# 두 값의 비율을 비교하기 위해 데이터 준비
sales_data = {
    'Category': ['Cancelled Sales', 'Valid Sales'],
    'Amount': [cancel_price_sum, valid_sales_price_sum]
}

sales_df = pd.DataFrame(sales_data)

plt.figure(figsize=(8, 6))

# 파이 차트로 비율 비교
plt.pie(sales_df['Amount'], labels=sales_df['Category'], autopct='%1.1f%%', colors=['red', 'green'], startangle=140)
plt.title('Comparison of Cancelled Sales vs Valid Sales')

# 그래프 출력
plt.axis('equal')  # 파이 차트를 원형으로 표시
plt.show()
```

### cancel 데이터의 가격 분포

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV 파일 불러오기
df = pd.read_csv('online_retail_II.csv.zip', compression='zip')

# 가격대 범위 및 레이블 설정
price_bins = [0, 5, 10, 20, 50, 100, 200, 500, 1000, 5000, df['Price'].max()]
price_labels = ['0-5', '5-10', '10-20', '20-50', '50-100', '100-200', '200-500', '500-1000', '1000-5000', '5000+']

# 가격대를 기준으로 'Price'를 'bins'로 분할
df['Price_Range'] = pd.cut(df['Price'], bins=price_bins, labels=price_labels, include_lowest=True)

# 가격대별 매출(송장 및 거래 요청)의 빈도 계산
price_range_counts = df['Price_Range'].value_counts().sort_index()

# 시각화 준비
plt.figure(figsize=(12, 6))

# 가격대별 매출 빈도 시각화
sns.barplot(x=price_range_counts.index, y=price_range_counts.values, palette='coolwarm')
plt.title('Frequency of Sales by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Frequency of Invoices/Transactions')
plt.xticks(rotation=45)

plt.show()
```

```python
# 가격 범위를 0-5, 5-10, 10-20, 20-50으로 설정
price_bins_limited = [0, 5, 10, 20, 50]
price_labels_limited = ['0-5', '5-10', '10-20', '20-50']

# 가격대를 기준으로 'Price'를 제한된 범위로 분할
df['Price_Range_Limited'] = pd.cut(df['Price'], bins=price_bins_limited, labels=price_labels_limited, include_lowest=True)

# 가격대별 매출(송장 및 거래 요청)의 빈도 계산 (제한된 범위)
price_range_counts_limited = df['Price_Range_Limited'].value_counts().sort_index()

# 결과를 표로 만들어 출력합니다.
price_range_counts_limited_df = price_range_counts_limited.reset_index()
price_range_counts_limited_df.columns = ['Price Range', 'Frequency of Invoices']

price_range_counts_limited_df
```
